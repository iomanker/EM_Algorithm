```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw_curve(bins, mu, sigma, color='r'):
    plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
                   np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
             linewidth=2, color=color)
```

```python
x1 = np.random.normal(160, 5, 100)
```

```python
x2 = np.random.normal(180, 5, 100)
```

```python
count, bins, ignored = plt.hist(x1, 30, facecolor='g', density=True)
draw_curve(bins,x1.mean(),np.std(x1))
plt.show()
```

```python
count, bins, ignored = plt.hist(x2, 30, facecolor='b', density=True)
draw_curve(bins,x2.mean(),np.std(x2))
plt.show()
```

```python
x_mix = np.concatenate([x1,x2])
```

```python
count, bins, ignored = plt.hist(x_mix, 30, facecolor='linen', density=True)
draw_curve(bins,x1.mean(),np.std(x1), 'g')
draw_curve(bins,x2.mean(),np.std(x2), 'b')
x1_curve = 1/(np.std(x1) * np.sqrt(2 * np.pi)) * \
               np.exp( - (bins - x1.mean())**2 / (2 * np.std(x1)**2) )
x2_curve = 1/(np.std(x2) * np.sqrt(2 * np.pi)) * \
               np.exp( - (bins - x2.mean())**2 / (2 * np.std(x2)**2) )
plt.plot(bins, x1_curve + x2_curve,
         linewidth=2, color='r')
# draw_curve(bins,x_mix.mean(),np.std(x_mix))
plt.show()
```
